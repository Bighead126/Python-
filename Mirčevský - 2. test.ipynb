Vytvořte aplikaci, do které napíšete 5 vstupů.
Mezi vstupy bude minimálně 1 číslo a všechen text bude o minimální délce 5 znaků. Žádný ze vstupů nesmí obsahovat kombinaci čísla a textu.   (viz cvičení na githubu: isdigit/isalpha) PS: žádný ze vstupů si nemusíte pamatovat. Stačí pouze vypsat 5 validních vstupních hodnot)


````python
#TODO
pocet = 0

while pocet < 5:
    vstup = input("Zadejte číslo nebo slovo: ")

    if vstup.isdigit():
        print("zadali jste:", vstup)
        pocet += 1

    elif vstup.isalpha() and len(vstup) >= 5:
        print("zadali jste:", vstup)
        pocet += 1

    else:
        print("Zadejte jiný vstup, není splněná podmínka")
````

Poté vypočtěte průměr pro každého žáka a vypište ve formátu:  
„Jméno žáka“ : „průměr“


````python
#TODO
for i in range(3):
    jmeno = input("Zadejte jméno: ")
    
    if jmeno.lower() == "konec":
        break

    soucet = 0
    for j in range(5):
        znamka = float(input("Zadejte známku: "))
        soucet += znamka

    prumer = soucet / 5
    print(f"{jmeno}: {prumer}")

print("Konec")

````
