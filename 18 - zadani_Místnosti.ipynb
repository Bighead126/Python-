**Vytvořte Třídu Room, která bude obsahovat atributy:**


1. název
2. seznamExitu
3. poklad
4. nepritel

**Vytvořte metody:**


1.   addExit, getNazev, getExity, getPoklad, getNepritel, toString
2.   setNazev





```python
#TODO
class Room:
    def __init__(self, nazev):
        self.nazev = nazev
        self.seznam_exitu = []
        self.poklad = None
        self.nepritel = None
    def addExit(self, exit):
        self.seznam_exitu.append(self)
    def setPoklad(self, poklad):
        self.poklad = poklad
    def seznamToString(self):
        vystup = "Vychody z mistnosti " + self.nazev + ": "
        for exit in self.seznam_exitu:
            vystup += exit.nazev + " "
        return vystup
    
    def __str__(self):
        if self.poklad is not None and self.nepritel is not None:
            return f"Mistnost: {self.nazev}, Poklad: {self.poklad}, Nepritel: {self.nepritel}"
        elif self.poklad is not None:
            return f"Mistnost: {self.nazev}, Poklad: {self.poklad}"
        elif self.nepritel is not None:
            return f"Mistnost: {self.nazev}, Nepritel: {self.nepritel}"
        
    def add_exit(self, exit_room):
        self.seznam_exitu.append(exit_room)

    def getSeznamExitu(self):
        text = ""
        for exit in self.seznam_exitu:
            text += exit.nazev + " \n"
        return text
    
```

**Zde je vytvořená jednoduchá mapa s výpisy**

```python
room1 = Room("Vstup do jeskyně")
room2 = Room("Jezírko")
room3 = Room("Předsíň draka")
room4 = Room("Dračí slůj")
room5 = Room("Vyhlídka")

room1.addExit(room5)
room5.addExit(room1)
room1.addExit(room2)
room1.addExit(room4)
room2.addExit(room1)
room4.addExit(room1)
room4.addExit(room3)
room4.addExit(room5)
room3.addExit(room4)
room5.addExit(room4)

aktualni = room1
print(aktualni.getSeznamExitu())
print(aktualni.getNepritel())
print(aktualni.getPoklad())
print(aktualni)

#chci se přesunout tak si změním aktualní na jeden z východu
aktualni = aktualni.seznamExitu[0]
print(aktualni)
#aktualni = aktualni.seznamExitu[2] # již nemá 3 exit --> skončí chybou
#print(aktualni)
aktualni = aktualni.seznamExitu[1]
print(aktualni)
```
