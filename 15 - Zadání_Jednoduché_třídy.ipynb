Vytvořte třídu Člověk, která bude mít atributy:


1.   jméno:str
2.   příjmení:str
3.   telefon:int
4.   email:str
5. adresa:str
6. pojišťovna:str

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše o této osobě




````python
class Člověk:
    def __init__(self, jméno:str, přijmení:str, telefon:str, email:str, adresa:str, pojistovna:str):
        self.jméno = jméno
        self.prijimení = přijmení
        self.telefon = telefon
        self.email = email
        self.adresa = adresa
        self.pojistovna = pojistovna

    def get_jmeno(self):
        return self.jméno
    def set_jmeno(self, nove_jmeno):
        self.jméno = nove_jmeno
        print("Jméno bylo změněno")
    
    def get_prijimeni(self):
        return self.prijimení
    def set_prijimeni(self, nove_prijimeni):
        self.prijimení = nove_prijimeni
        print("Příjmení bylo změněno")
    
    def get_telefon(self):
        return self.telefon
    def set_telefon(self, novy_telefon):
        self.telefon = novy_telefon
        print("Telefonní číslo bylo změněno")

    def get_email(self):
        return self.email
    def set_email(self, novy_email):
        self.email = novy_email
        print("E-mailová adresa byla změněna")

    def get_adresa(self):
        return self.adresa
    def set_adresa(self, nova_adresa):
        self.adresa = nova_adresa
        print("Adresa byla změněna")

    def get_pojistovna(self):
        return self.pojišťovna
    def set_pojistovna(self, nova_pojistovna):
        self.pojišťovna = nova_pojistovna
        print("Pojišťovna byla změněna")


    def __str__(self):
        return f"""
        Jméno: {self.jméno} 
        Příjmení: {self.prijimení}
        Telefon: {self.telefon} 
        E-mail: {self.email}
        Adresa: {self.adresa} 
        Pojišťovna: {self.pojistovna}
        """

clovek1 = Člověk("Aneta", "Koudelková", 123456789,"anet@spskladno.cz","jahodova 12","VZP")
clovek2 = Člověk("asd","ADASD",4598794564156,"andejdf@spskladno.cz", "asdasd 12","VZDR")
print(clovek1)

````

upravte třídu Člověk tak, abyste ošetřili následující atributy:


1.   jméno:str
2.   příjmení:str
3.   telefon:int (maximálně 9 znaků)
4.   email:str (musí obsahovat @)
5. adresa:str
6. pojišťovna:str

Upravte metody, které budou moc měnit tyto atributy i s aplikovanými omezeními.




````python
class Člověk:
    def __init__(self, jmeno: str, prijmeni: str, telefon: int,
                 email: str, adresa: str, pojistovna: str):

        self.set_jmeno(jmeno)
        self.set_prijmeni(prijmeni)
        self.set_telefon(telefon)
        self.set_email(email)
        self.set_adresa(adresa)
        self.set_pojistovna(pojistovna)

    
    def get_jmeno(self):
        return self._jmeno

    def set_jmeno(self, nove_jmeno):
        if not isinstance(nove_jmeno, str) or nove_jmeno.strip() == "":
            raise ValueError("Jméno musí být neprázdný řetězec.")
        self._jmeno = nove_jmeno

   
    def get_prijmeni(self):
        return self._prijmeni

    def set_prijmeni(self, nove_prijmeni):
        if not isinstance(nove_prijmeni, str) or nove_prijmeni.strip() == "":
            raise ValueError("Příjmení musí být neprázdný řetězec.")
        self._prijmeni = nove_prijmeni

   
    def get_telefon(self):
        return self._telefon

    def set_telefon(self, novy_telefon):
        if not isinstance(novy_telefon, int):
            raise ValueError("Telefon musí být číslo.")
        if len(str(novy_telefon)) > 9:
            raise ValueError("Telefonní číslo může mít maximálně 9 číslic.")
        self._telefon = novy_telefon

   
    def get_email(self):
        return self._email

    def set_email(self, novy_email):
        if not isinstance(novy_email, str) or "@" not in novy_email:
            raise ValueError("Email musí obsahovat znak @.")
        self._email = novy_email

    
    def get_adresa(self):
        return self._adresa

    def set_adresa(self, nova_adresa):
        if not isinstance(nova_adresa, str) or nova_adresa.strip() == "":
            raise ValueError("Adresa nesmí být prázdná.")
        self._adresa = nova_adresa

   
    def get_pojistovna(self):
        return self._pojistovna

    def set_pojistovna(self, nova_pojistovna):
        if not isinstance(nova_pojistovna, str) or nova_pojistovna.strip() == "":
            raise ValueError("Pojišťovna nesmí být prázdná.")
        self._pojistovna = nova_pojistovna

    def __str__(self):
        return (
            f"Jméno: {self._jmeno}\n"
            f"Příjmení: {self._prijmeni}\n"
            f"Telefon: {self._telefon}\n"
            f"E-mail: {self._email}\n"
            f"Adresa: {self._adresa}\n"
            f"Pojišťovna: {self._pojistovna}"
        )



clovek1 = Člověk("Aneta", "Koudelková", 123456789,
                 "anet@spskladno.cz", "Jahodová 12", "VZP")
print(clovek1)


````
